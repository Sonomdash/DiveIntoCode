Problem 1 Creating random numbers
    
    
random.multivariate_normal(mean, cov, size=None, check_valid='warn', tol=1e-8)
    

```python
import numpy as np
np.random.seed(0) 
mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
import matplotlib.pyplot as plt
x = np.random.multivariate_normal(mean, cov, 500)
print(x.shape)
print(x)
```

Problem 2 Visualization by scatter plot

```python
import matplotlib.pyplot as plt
y= x[:,0]
z= x[:,1]
plt.scatter(y,z, alpha=0.9,label="scatter")
plt.xlabel("z")
plt.ylabel("y")
plt.legend(loc='upper left')
plt.show()
```

Problem 3 Visualization by histogram

```python
plt.title('Histogram x1')
plt.xlabel('x1')
plt.ylabel('frequency')
plt.hist(y,  facecolor='blue', alpha=0.5)
plt.xlim(-5, 2)

plt.show()
plt.title('Histogram x2')
plt.xlabel('x2')
plt.ylabel('frequency')
plt.hist(z,  facecolor='red', alpha=0.5)
plt.xlim(-5, 2)
plt.show()
```
