Create dart throw simulation

Problem 1 Choosing a dart mat

```python
import numpy as np
while True:
    random_number = np.random.choice(6)
    if(random_number) != 0 :
        break
        
print(random_number)
```

Problem 2 Throw darts and record the place

```python
mean = (0, 0)
cov=[[0.01,0],[0,0.01]]
x = np.random.multivariate_normal(mean, cov)
print(x)
```

Problem 3 Throw a lot of darts and record them in a state of being tied to the number

```python
create = {}

for x in range(1 , 1001):
    target_number = np.random.choice(6) + 1
    create[str(x)] = target_number

print(create)
multivarti={}
for x in range(1 , 1001):
    y = np.random.multivariate_normal(mean, cov)
    multivarti[str(x)] = y
print(multivarti)
```

Problem 4 Drawing the position of the darts

```python
import matplotlib.pyplot as plt

figure, axes = plt.subplots() 

boolean = True    
for x in range (10,0,-1):
    if boolean == True :
        circle = plt.Circle(( 0.0 , 0.0 ), x, color = "blue" ) 
        boolean = False
       
    else :
        circle = plt.Circle(( 0.0 , 0.0 ), x, color = "red" ) 
        boolean = True
    axes.add_artist( circle ) 


axes.set_xlim([-15,15])
axes.set_ylim([-15,15])
plt.title( 'Darts scoreboard' ) 
plt.show()
```
