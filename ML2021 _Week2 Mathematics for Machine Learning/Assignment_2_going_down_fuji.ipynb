Problem 1 Visualization of data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
csv_path = "mtfuji_data.csv" # Specify the file name (path)
np.set_printoptions(suppress=True) 
# Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)

x = fuji[:,3:4]
y = fuji[:,0:1]
plt.title("Mt Fuji")
plt.xlabel("position")
plt.ylabel("elevation(m)")
plt.plot(y,x,linewidth=3 , color = 'red')
pd.DataFrame(fuji,
                   columns=['point', ' Latitude ', 'Longitude','Elevation','Distance'])
```

Problem 2 Create a function to calculate the slope of a point



```python
array = np.array(fuji[:,3:4])
array_1 = np.array( fuji[:,0:1])
gradient = np.array((array[1:]-array[:-1]) / (array_1[1:] - array[:-1]))
print(gradient)
def calculate_slope(point):
    change_in_elevation = fuji[:,0][point] - fuji[:,0][point-1]
    change_in_position = fuji[:,3][point] - fuji[:,3][point-1]

    slope = change_in_position/change_in_elevation
    return slope
print(calculate_slope(0))
```

<h3>Problem 3 Create a function to calculate the destination<h3>

Destination point = current location-Î± Ã— Gradient at the current location
    
    
Î± - is the hyper-parameters to take any value greater than 0. In the field of machine learning, it is called the learning rate

```python

def destination_point_function(current_location , alpha , gradient ):
   
    destination_point = current_location - alpha  * gradient
    if destination_point <=0:
        return 0
    else :
        return round(destination_point)
    return destination_point
alpha = 0.2
current_point = 50
gradient = calculate_slope(current_point)
destination_point_function(current_point,alpha,gradient)


```

<h3>Problem 4 Creating a function to go down the mountain<3>

```python

alp = 0.2
point = 136
def calculate_going_down(point,alp):
    slope_1 = calculate_slope(point)
    position = [point]
  
    for x in range(fuji.size):
        point = int(destination_point_function(point,alp,slope_1))
        if position[-1]== point:
            break
        else :
            position.append(point)
    return position
calculate_going_down(point,alp)    
    
```

<h3>Problem 5 Visualization of the descent process<h3>

```python
position_array = calculate_going_down(136,0.2)
print("The height above the changed points ")
print([fuji[position_2 ,3] for position_2 in position_array])
plt.title("Mt Fuji")
plt.xlabel("position")
plt.ylabel("elevation(m)")
plt.plot(fuji[:,0],fuji[:,3],linewidth=3 , color = 'red')
plt.scatter(position_array, [fuji[position_2 ,3] for position_2 in position_array],c='black',label = 'elevation')

figure =plt.figure(figsize=(15,5))
plt_2 = figure.add_subplot(1,2,2)
plt_2.set_title("Slope")
plt_2.set_xlabel("slope")
plt_2.set_ylabel("elevation(m)")
plt_2.plot(position_array, [fuji[position_2 ,3] for position_2 in position_array],c='black',label = 'elevation')
plt_2.plot(position_array, [calculate_slope(position_2 )for position_2 in position_array],c='green',label = 'slope')
plt_2.legend()
```

```python
def display_graph_give_points(points):
    
    alpha_3 = 0.2
    for  x in range(points.size):
        position_arr = calculate_going_down(points[x],alpha_3)
        figure =plt.figure(figsize=(15,5))
        plt_1 = figure.add_subplot(1,2,1)
        plt_1.set_title("Mt Fuji")
        plt_1.set_xlabel("position")
        plt_1.set_ylabel("elevation(m)")
        plt_1.plot(fuji[:,0],fuji[:,3],linewidth=3 , color = 'red')
        plt_1.scatter(position_arr, [fuji[position_2 ,3] for position_2 in position_arr],c='blue',label = 'elevation')
      
    
        plt_2 = figure.add_subplot(1,2,2)
        plt_2.set_title("Slope")
        plt_2.set_xlabel("slope")
        plt_2.set_ylabel("elevation(m)")
        plt_2.plot(position_arr, [fuji[position_2 ,3] for position_2 in position_arr],c='blue',label = 'elevation')
        plt_2.plot(position_arr, [calculate_slope(position_2 )for position_2 in position_arr],c='green',label = 'slope')
        plt_2.legend()
points = np.array([100,120,130])
display_graph_give_points(points)
```
