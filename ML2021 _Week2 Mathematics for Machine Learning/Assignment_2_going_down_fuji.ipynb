Problem 1 Visualization of data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
csv_path = "mtfuji_data.csv" # Specify the file name (path)
np.set_printoptions(suppress=True) 
# Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)

x = fuji[:,3:4]
y = fuji[:,0:1]
plt.title("Mt Fuji")
plt.xlabel("position")
plt.ylabel("elevation(m)")
plt.plot(y,x,linewidth=3 , color = 'red')
pd.DataFrame(fuji,
                   columns=['point', ' Latitude ', 'Longitude','Elevation','Distance'])
```

Problem 2 Create a function to calculate the slope of a point



```python
array = np.array(fuji[:,3:4])
array_1 = np.array( fuji[:,0:1])
gradient = np.array((array[1:]-array[:-1]) / (array_1[1:] - array[:-1]))

print(gradient)
```

<h3>Problem 3 Create a function to calculate the destination<h3>

Destination point = current location-α × Gradient at the current location
    
    
α - is the hyper-parameters to take any value greater than 0. In the field of machine learning, it is called the learning rate

```python
current_location = fuji[0:,1:4]
print("Current location Latitude , Longitude, Elevation,distance\n",current_location[1:10])
α = 0.2
def Destination_point_function( current_location ):
   
    destination_point = current_location - α  * gradient
    return destination_point
print("Next destination,Latitude , Longitude, Elevation,distance ")
print(Destination_point_function(current_location[1:]))
```

Problem 4 Creating a function to go down the mountain

```python
arr = Destination_point_function(fuji[1:])
pd.DataFrame(arr,
                   columns=['point', ' Latitude ', 'Longitude','Elevation','Distance'])
```
