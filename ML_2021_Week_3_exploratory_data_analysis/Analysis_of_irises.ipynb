Problem 6 Creating a diagram

Matplotlib - Violin Plot
Violin plots are similar to box plots, except that they also show the probability density of the data at different values.

Matplotlib - Box Plot
A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. 



```python
import matplotlib.pyplot as plt

def  column_different_ways(x,y,values):
    i=0
    array=[]
    for _ in union_x_y.index: 
        array.append(union_x_y.loc[i][values])
        i+=1
    data = []
    for index in range(len(array)):
        if array[index] >= x and array[index] < y:
            data.append(array[index])
            
    return len(data)
    
bet_2_3 = column_different_ways(2.0,3.0,1)
bet_3_4 = column_different_ways(3.0,4.0,1)
bet_4 = column_different_ways(4.0,5.0,1)
def percentages(value):
    length = len(union_x_y.index)
    val = (value *100)/length
    return val
print("                    Sepal width pie ")

labels = 'Values between 2.0 and 3.0', 'Values between 3.0 and 4.0', '4.0 max values'
sizes = [percentages(bet_2_3), percentages(bet_3_4), percentages(bet_4)]
explode = (0, 0.1, 0) 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.



def values(x ):
    i =0
    array = []
    for _ in union_x_y.index: 
        array.append(union_x_y.loc[i][x])
        i+=1
    return array
sepal_length_data =values(0)
sepal_width_data = values(1)
petal_length_data = values(2)
petal_width_data = values(3)

data = [sepal_length_data, sepal_width_data, petal_length_data, petal_width_data]
  
fig = plt.figure(figsize =(10, 7))
  
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
  
# Creating plot
bp = ax.boxplot(data)
  
# show plot


fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data)

```
